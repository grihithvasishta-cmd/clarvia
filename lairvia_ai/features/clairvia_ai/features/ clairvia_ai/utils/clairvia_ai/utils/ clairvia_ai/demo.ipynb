# Clairvia AI Demo Notebook

This notebook demonstrates a simple end-to-end run of the Clairvia AI pipeline on sample text.

```python
from clairvia_ai.main import main
from clairvia_ai.utils.file_utils import FileHandler
pipeline = main()
fh = FileHandler()
print('Pipeline initialized')
```

```python
sample_text = '''\nPhysics is the natural science that studies matter, its motion, and behavior through space and time, and that studies the related entities of energy and force.\nNewton's second law: F = m a.\nThe law of universal gravitation was formulated by Isaac Newton in 1687.\nEnergy is the ability to do work.\nFor example, a falling object converts potential energy into kinetic energy.\n'''
fh.write_text('temp/sample.txt', sample_text)
print('Sample text written')
```

```python
text = fh.read_text('temp/sample.txt')
cleaned = pipeline.text_cleaner.clean(text)
structure = pipeline.structure_builder.build(cleaned)
depth_notes = pipeline.depth_notes_gen.generate(structure)
cheat_notes = pipeline.cheat_notes_gen.generate(structure)
flashcards = pipeline.flashcard_gen.generate(structure)
mock_test = pipeline.mock_test_gen.generate(structure)
print('Structure chapters:', [c['name'] for c in structure.get('chapters', [])])
print('Depth notes keys:', list(depth_notes.keys()))
print('Cheat notes keys:', list(cheat_notes.keys()))
print('Flashcards sample:', flashcards[:2])
print('Mock test summary:', mock_test['total_questions'])
```

You can extend this demo by replacing the sample text with PDFs, audio transcriptions, or YouTube transcripts.
